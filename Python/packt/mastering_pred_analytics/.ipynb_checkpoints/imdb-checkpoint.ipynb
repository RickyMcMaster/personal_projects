```python
import pandas as pd
import numpy as np

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# MPL Settings
import matplotlib as mpl,  matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

mpl.rcParams["font.family"] = "Arial"
mpl.rcParams['figure.dpi'] = 300
%matplotlib inline

grey = '#808080'
darkgrey = '#505050'
# (http://www.color-hex.com/color/505050)

mpl.rcParams['axes.linewidth'] = 0.3
mpl.rcParams['axes.edgecolor'] = grey
mpl.rcParams['xtick.color'] = grey
mpl.rcParams['ytick.color'] = grey
mpl.rcParams['text.color'] = darkgrey
mpl.rcParams['axes.labelcolor'] = darkgrey
```

```python
# Data Import
imdb_ratings = pd.read_csv('MasteringPredictiveAnalyticswithPython_Code/B04881_02_code/movies.csv')
```

```python
# Basic Features
imdb_ratings.head()
imdb_ratings.columns
imdb_ratings.shape
imdb_ratings.describe()
imdb_ratings.dtypes
```

# Basic Exploration

```python
imdb_ratings['year'].value_counts().sort_index(ascending=False)
```

```python
# R-Rated 21st Century
imdb_ratings[(imdb_ratings.year > 1999) & (imdb_ratings.mpaa ==
'R')].head()
```

```python
# Ratings Annual Stats
imdb_ratings.groupby('year').rating.aggregate([np.mean,np.std])
```

```python
imdb_ratings.groupby('year').rating.aggregate([np.mean,np.std]).plot()
```

```python
imdb_ratings[['title', 'Action','Animation','Comedy','Drama','Documentary','Romance']].head(20)
```

```python
imdb_ratings['genre']=imdb_ratings[['Action','Animation','Comedy','Drama','Documentary','Romance']].idxmax(1)
```

```python
imdb_ratings[['title', 'Action','Animation','Comedy','Drama','Documentary','Romance', 'genre']].head(20)
```

```python
# But, if there is more than one category then it seems like the first one in column order is assigned
```

```python
# Colour map
genres_map = {"Action": 'red', "Animation": 'blue', "Comedy":
'yellow', "Drama": 'green', "Documentary": 'orange', "Romance": 'purple'}
imdb_ratings['genre_color'] = imdb_ratings['genre'].apply(lambda x:
genres_map[x])
imdb_ratings['genre_color'].head()
```

```python
# Pivot for year, genre and rating
pd.pivot_table(imdb_ratings,values='rating',index='year',columns=['genre'],aggfunc=np.mean).tail(20)
```

```python
pd.pivot_table(imdb_ratings,values='rating',index='year',columns=['genre'],aggfunc=np.mean).tail(20).plot()
```

# Visualisations

```python
imdb_ratings.length.plot()
```

```python
plot1 = imdb_ratings.groupby('genre').length.plot(kind='kde',xlim=(0,120),legend='genre')
plot1[0].set_xlabel('Number of Minutes')
plot1[0].set_title('Distribution of Films by Runtime Minutes')
```
