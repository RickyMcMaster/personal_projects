```python
def readPostbank(src_file, skipno):
    print('Reading {}'.format(src_file))
#   Find row number of main dataset and skip previous
    with open(src_file, 'r') as o:
        myData = csv.reader(o, delimiter=';')
        skipno = 0
        for row in myData:
            for c in row:
                if 'Wertstellung' in c:
                    skipno += myData.line_num-1
    return pd.read_csv(src_file, sep=';'\
    , encoding='utf-8', skiprows=skipno, parse_dates=['Buchungsdatum','Wertstellung'], dayfirst=True)
```

```python
spreadsheet = '1_nW4cYU1Jvu7K0wwo0Jxt1730TWc_dSmnt-GQC2PHpQ'
wks_name = str(current_mth)

df_budget = g2d.download(spreadsheet, wks_name, col_names = True, row_names = False)
df_budget['Budget_Month'] = (current_mth)
df_budget.Budget_Amount = df_budget.Budget_Amount.astype(float)
```

```python
df_budget.head()
df_budget.info()
df_budget.columns
```

```python
df_budget = df_budget[['Budget_Type', 'Budget_Category', 'Budget_Sub-Category',
       'Budget_Amount']]
df_budget.head()
```

```python
df_budg_gpd = pd.DataFrame(df_budget.groupby(['Budget_Type', 'Budget_Category'], as_index=False)['Budget_Amount'].sum())
df_budg_gpd = df_budg_gpd.sort_values(by='Budget_Amount', ascending=False)
df_budg_gpd
```

```python
df_mgd = df_current[df_current['Category']!='Wages'].merge(df_budg_gpd, left_on=['Category'], right_on=['Budget_Category'], how='outer')
df_mgd['Budget_Net'] = df_mgd.Amount_EUR + df_mgd['Budget_Amount']
df_mgd['Budget_Used'] = ((df_mgd.Amount_EUR*-1) / df_mgd['Budget_Amount'])
cols = ['Budget_Type', 'Category','Budget_Category', 'Amount_EUR', 'Budget_Amount', 'Budget_Net', 'Budget_Used']
df_mgd = df_mgd.sort_values(by='Budget_Used',ascending=False)
df_mgd[cols].head()
```

```python
# df_mgd[['Budget_Category', 'Amount_EUR', 'Budget_Amount']].plot.bar()
df_mgd['Amount_EUR_Abs'] = df_mgd['Amount_EUR'].abs()
ax = pd.pivot_table(df_mgd, index=['Budget_Category'], values=['Amount_EUR_Abs', 'Budget_Amount'], aggfunc='sum')\
.sort_values(by='Amount_EUR_Abs').plot(kind='barh', colormap='Dark2')
rcParams['figure.figsize'] = 20, 15
rcParams['figure.dpi'] = 300
ax.set_xlabel("")
L = ax.legend()
plt.setp(L.texts, family='Arial')
plt.xticks(fontsize=18, rotation=45, fontname='Arial')
plt.yticks(fontsize=18, fontname='Arial')
ax.set_xlabel("Amount €", fontname="Arial", fontsize=18)
ax.set_ylabel("Category", fontname="Arial", fontsize=18)
ax.set_title("Spending vs Budget Current Month", fontname='Arial', fontsize=20)
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "€{:,}".format(int(x))))
ax.grid(False, axis='y')
plt.show()
```
