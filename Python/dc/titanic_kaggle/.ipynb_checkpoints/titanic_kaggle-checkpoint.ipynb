## First Machine Learning Model

```python
# Store target variable of training data in a safe place
survived_train = df_train.Survived

# Concatenate training and test sets
data = pd.concat([df_train.drop(['Survived'], axis=1), df_test], sort=False)
data.info()
```

```python
# Impute missing numerical variables (using median, because data is skewed)
data['Age'] = data.Age.fillna(data.Age.median())
data['Fare'] = data.Fare.fillna(data.Fare.median())

# Check out info of data
data.info()
```

```python
# Convert Sex to Numeric
data = pd.get_dummies(data, columns=['Sex'], drop_first=True)
data.head()
```

```python
# Select columns and view head
data = data[['Sex_male', 'Fare', 'Age','Pclass', 'SibSp']]
data.head()
data.info()
```

```python
# Split into test and train again
data_train = data.iloc[:891]
data_test = data.iloc[891:]
```

```python
X = data_train.values
test = data_test.values
y = survived_train.values
```

```python
# Instantiate model and fit to data
clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(X, y)
```

```python
# Make predictions and store in 'Survived' column of df_test
Y_pred = clf.predict(test)
df_test['Survived'] = Y_pred
```

```python
df_test[['PassengerId', 'Survived']].to_csv('data/predictions/1st_dec_tree.csv', index=False)
```
