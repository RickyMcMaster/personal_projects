```python
df = pd.read_excel('NYC Boroughs & Neighbourhoods.xlsx', sheet_name='Data')
```

```python
df
```

```python
df_fin = (pd.melt(df, id_vars=['Borough', 'Neighborhood'], value_vars=['Zip1', 'Zip2','Zip3','Zip4','Zip5','Zip6', 'Zip7', 'Zip8','Zip9'], value_name='ZIP'))
```

```python
df_fin[df_fin.Neighborhood=='Bushwick and Williamsburg'].head(100)
```

```python
df_fin = df_fin.dropna()
df_fin = df_fin[['Borough', 'Neighborhood', 'ZIP']]
```

```python
df_fin.to_excel('NYC_BorNeighZIP.xlsx', index=False)
```
